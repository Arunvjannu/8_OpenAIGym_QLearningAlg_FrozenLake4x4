```python
import numpy as np
import gym
import random
import matplotlib.pyplot as plt
```

```python
env = gym.make("FrozenLake-v1", map_name="8x8", is_slippery=False)
```

```python
action_space_size = env.action_space.n  # 4 actions: Left, Down, Right, Up
state_space_size = env.observation_space.n  # 64 states (8x8 grid)
q_table = np.zeros((state_space_size, action_space_size))
```

```python
num_episodes = 25000
max_steps_per_episode = 300

learning_rate = 0.1
discount_rate = 0.99

exploration_rate = 1  # Start with full exploration
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate = 0.0003
```

```python
rewards_all_episodes = []

for episode in range(num_episodes):
    state = env.reset()[0]  # Reset environment for each episode
    done = False
    rewards_current_episode = 0

    for step in range(max_steps_per_episode):
        # Explore or Exploit
        exploration_threshold = random.uniform(0, 1)
        if exploration_threshold > exploration_rate:
            action = np.argmax(q_table[state, :])  # Exploit
        else:
            action = env.action_space.sample()  # Explore

        new_state, reward, done, truncated, info = env.step(action)

        # Q-learning formula
        q_table[state, action] = q_table[state, action] * (1 - learning_rate) + \
                                 learning_rate * (reward + discount_rate * np.max(q_table[new_state, :]))

        state = new_state
        rewards_current_episode += reward

        if done:
            break

    # Decay exploration rate
    exploration_rate = min_exploration_rate + \
        (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate * episode)

    rewards_all_episodes.append(rewards_current_episode)
```

```python
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

moving_avg_rewards = moving_average(rewards_all_episodes, 1000)

plt.figure(figsize=(10,5))
plt.plot(moving_avg_rewards)
plt.xlabel("Episode")
plt.ylabel("Average Reward (over last 1000 episodes)")
plt.title("Agent's Learning Progress on 8x8 FrozenLake")
plt.grid()
plt.show()
```

```python
print("âœ… Testing the trained agent on 8x8 FrozenLake:")

for episode in range(5):
    state = env.reset()[0]
    done = False
    print(f"\nEPISODE {episode + 1}\n")
    step_count = 0

    for step in range(max_steps_per_episode):
        env.render()
        action = np.argmax(q_table[state, :])  # Always choose best action
        new_state, reward, done, truncated, info = env.step(action)
        state = new_state
        step_count += 1
        if done:
            if reward == 1:
                print(f"\nðŸŽ‰ Reached the goal in {step_count} steps!")
            else:
                print(f"\nðŸ˜ž Fell into a hole or failed.")
            break

env.close()
```
